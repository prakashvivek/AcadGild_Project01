```python
#17. Build a dataset with every possible (UserID, ProductID) pair (cross join)
```

```python
userDF = users.loc[:,'UserID':'UserID']
```

```python
productDF = products.loc[:,'ProductID':'ProductID']
```

```python
index = pd.MultiIndex.from_product([userDF.UserID, productDF.ProductID], names = ["UserID", "ProductID"])
userIDproductIDdf = pd.DataFrame(index = index).reset_index()
print(userIDproductIDdf)
```

```python
#18. Determine how much quantity of each product was purchased by each user
```

```python
#Creating the Dataframe by joining Users and transactions
user_product_Transdf = pd.merge(userIDproductIDdf, transactions, how='left', on=['UserID','ProductID'])

#Drop the unwanted column from the result dataframe
user_product_TransdfResult = user_product_Transdf.drop(['TransactionID', 'TransactionDate'], axis=1)

#Set NaN values to zero(0) from result dataframe
user_product_TransdfResult1 = user_product_TransdfResult.fillna(0)

print(user_product_TransdfResult1) # Print the fnal result
```

```python
#20. Join each user to his/her first occuring transaction in the transactions table
```

```python
# first select the first transaction for each user
firstOccur_transactions = transactions.groupby('UserID').first().reset_index()

#drop the unwanted column(s)
firstOccur_transactions = firstOccur_transactions.drop(['TransactionDate'], axis=1)

firstOccur_transactionsByEachUser = pd.merge(users, firstOccur_transactions, how='left', on=['UserID'])
print(firstOccur_transactionsByEachUser)
```
